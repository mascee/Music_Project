# CatBoost  
CatBoost is short for Categorical Boosting, it is a gradient-boosting algorithm developed by Yandex. It is designed to handle categorical features without the need for extensive preprocessing. It is Known for speed and the ability to withstand rigorous testing. 

The module was suggested through a ChatGPT search about ML model to test music files for genre [January 2025]. 
Code references: [https://github.com/catboost/catboost], [https://www.geeksforgeeks.org/catboost-tree-parameters/] , [https://www.datacamp.com/tutorial/catboost]

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from catboost import CatBoostClassifier, Pool
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = pd.read_csv ("CSV\mldata_v2.csv")
```

```python
data.head(11)
```

```python
data.columns
```

```python
# Drop irrelevant columns
data = data.drop(columns=['filename', 'length'])
data.head()
```

```python
# Define target vector "label"
y = data["label"].to_numpy()
y[:5]
```

```python
# Split into features and target
X = data.drop(columns=['label'])  
y = data['label'] 
```

```python
# Encode target labels if they are categorical
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
# Train CatBoost Model
catboost_model = CatBoostClassifier(
    iterations=1000,       
    learning_rate=0.1,    
    depth=6,                  
    random_seed=42,     
    verbose=50     # output the training process every 50 iterations       
)

catboost_model.fit(X_train, y_train)
```

```python
# Evaluate the Model
y_pred = catboost_model.predict(X_test)
print("Accuracy on Test Set:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
